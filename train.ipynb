```python
import os
import shutil
import cv2

# Paths
train_dir = 'dataset/Train'
test_dir = 'dataset/Test'

output_train_dir = 'yolo_dataset/train'
output_test_dir = 'yolo_dataset/test'

os.makedirs(os.path.join(output_train_dir, 'images'), exist_ok=True)
os.makedirs(os.path.join(output_train_dir, 'labels'), exist_ok=True)

os.makedirs(os.path.join(output_test_dir, 'images'), exist_ok=True)
os.makedirs(os.path.join(output_test_dir, 'labels'), exist_ok=True)

# Define class mapping (for example: 'melanoma' -> 0, 'nevus' -> 1, etc.)
class_mapping = {
    'actinic keratosis': 0,
    'seborrheic keratosis': 1,
    'basal cell carcinoma': 2,
    'nevus': 3,
    'pigmented benign keratosis': 4,
    'vascular lesion': 5
}

# Function to process a directory (either train or test)
def process_directory(directory, output_dir):
    for class_name, class_id in class_mapping.items():
        class_folder = os.path.join(directory, class_name)

        if not os.path.exists(class_folder):
            continue

        for image_name in os.listdir(class_folder):
            if image_name.endswith('.jpg') or image_name.endswith('.png'):
                image_path = os.path.join(class_folder, image_name)

                # Read image to get image dimensions
                img = cv2.imread(image_path)
                if img is None:
                    print(f"Error reading {image_name}, skipping.")
                    continue

                height, width, _ = img.shape

                # Example of annotation: You should replace this with real bounding box data if available
                # For this example, we'll just create dummy annotations
                # Assuming only 1 object per image at the center
                # In real cases, you need to extract bounding boxes
                x_center = 0.5  # Normalize to 0-1
                y_center = 0.5  # Normalize to 0-1
                w = 0.5         # Normalize to 0-1 (50% width)
                h = 0.5         # Normalize to 0-1 (50% height)

                # Save label file in YOLO format
                label_file = os.path.join(output_dir, 'labels', image_name.replace('.jpg', '.txt').replace('.png', '.txt'))
                with open(label_file, 'w') as f:
                    f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {w:.6f} {h:.6f}\n")

                # Save image to corresponding folder
                shutil.copy(image_path, os.path.join(output_dir, 'images', image_name))

# Process train and test directories
print("Processing training data...")
process_directory(train_dir, output_train_dir)
print("Processing test data...")
process_directory(test_dir, output_test_dir)

print("✅ YOLO Dataset Conversion Finished!")
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the path to your data.yaml file
data_yaml_path = 'data.yaml'

# Initialize the YOLO model (make sure the checkpoint exists or change to a correct one like 'yolov8n-cls.pt' if it's classification)
model = YOLO('yolo11n.pt')  # Make sure your model path is correct and ends with '.pt'

# Train the model
model.train(data=data_yaml_path, epochs=25, batch=12, imgsz=420, name='skin_cancer_model', exist_ok=True)
```

```python
# Save the trained model (optional)
model.save()

# Evaluate the model after training
results = model.val()

maps = results.maps  # Array of mAP50 scores per class
names = results.names  # Dictionary mapping class indices to class names

# Print per-class mAP50 scores (as approximate classification accuracy)
for idx, score in enumerate(maps):
    class_name = names.get(idx, f'class_{idx}')
    print(f"Class '{class_name}': {score * 100:.2f}% mAP50")
```
